## Merge the 12 months of sales data

```python

file_names = [file for file in os.listdir('Data_Source')]
merged_data_df = pd.DataFrame() 

for file_name in file_names:
    print('loading:',file_name)
    data = pd.read_csv('Data_Source/'+file_name)
    merged_data_df = pd.concat([merged_data_df, data], ignore_index=True)
```

```python
# quy mo la cua hang truyen thong nen k can dia chi
formatted_data_df = cleaned_data_df.drop("Purchase Address", axis='columns').reset_index(drop=True)
formatted_data_df.info()
```

```python

formatted_data_df = formatted_data_df.rename(columns={'Order ID':'Sale_ID','Quantity Ordered':'Quantity_Sold',
                                                      'Price Each':'Each_Price','Order Date':'Sale_Date'}).reset_index(drop=True)
formatted_data_df
```

```python

formatted_data_df['Sale_ID'],formatted_data_df['Quantity_Sold'],formatted_data_df['Each_Price']=formatted_data_df['Sale_ID'].astype('int64'),formatted_data_df['Quantity_Sold'].astype('int64'),formatted_data_df['Each_Price'].astype('float64')
```

```python

formatted_data_df['Sale_Date'] = pd.to_datetime(formatted_data_df['Sale_Date'])
formatted_data_df.info()
```

```python

formatted_data_df['Sales'] = formatted_data_df['Quantity_Sold']*formatted_data_df['Each_Price']
formatted_data_df
```

```python

formatted_data_df=formatted_data_df.sort_values(by='Sale_Date').reset_index(drop=True)
formatted_data_df
```

```python
formatted_data_df.describe()
```

```python

formatted_data_df.to_csv('Processed_Data.csv', index=False)
```

```python

stock_df = formatted_data_df.drop(['Sale_ID','Each_Price','Sale_Date','Sales'],axis='columns').reset_index(drop=True)
stocks_df = stock_df.groupby(['Product'],as_index=False)['Quantity_Sold'].sum()
stocks_df
```

```python

stocks_df.loc[:,'Quantity_Sold'] = 31000
stocks_df.rename(columns={'Quantity_Sold':'Stock_Quantity'}, inplace=True)
stocks_df
```

```python

stocks_df.to_csv('Stock_Quantity.csv', index=False)
```
